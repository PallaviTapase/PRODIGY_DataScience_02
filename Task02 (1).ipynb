```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set_style("whitegrid")
```

```python
# Load the dataset
df = pd.read_csv("train.csv")  # Use test.csv if needed

# Display first few rows
df.head()
```

```python
# Overview of dataset
df.info()

# Check missing values
df.isnull().sum()
```

```python
# Fill missing Age values with median
df["Age"].fillna(df["Age"].median(), inplace=True)

# Fill missing Embarked values with most common value
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)

# Drop Cabin column (too many missing values)
df.drop(columns=["Cabin"], inplace=True)

# Verify missing values are handled
df.isnull().sum()
```

```python
# Countplot for Survived column
plt.figure(figsize=(5,3))
sns.countplot(x="Survived", data=df, palette="coolwarm")
plt.title("Survival Count")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(x="Sex", y="Survived", data=df, palette="viridis")
plt.title("Survival Rate by Gender")
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.barplot(x="Pclass", y="Survived", data=df, palette="rocket")
plt.title("Survival Rate by Passenger Class")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Age"], bins=30, kde=True, color="blue")
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", linewidths=1)
plt.title("Feature Correlation Heatmap")
plt.show()
```
